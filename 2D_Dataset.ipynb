```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
def generate_dataset(separability, cluster_sizes):
    np.random.seed(42)

    # Generate Class 0
    class_0 = np.random.randn(cluster_sizes[0], 2) + np.array([0, 0])

    # Generate Class 1 with separability
    class_1 = np.random.randn(cluster_sizes[1], 2) + np.array([separability, separability])
    class_2 = np.random.randn(cluster_sizes[2], 2) + np.array([-separability, separability])

    # Combine data
    X = np.vstack((class_0, class_1,class_2))
    y = np.array([0] * cluster_sizes[0] + [1] * cluster_sizes[1]+[2] * cluster_sizes[2])

    return X, y
```

```python
def visualize_dataset(X, y, title="Dataset"):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 0, 0], X[y == 0, 1], c='red', label='Class 0')
    plt.scatter(X[y == 1, 0], X[y == 1, 1], c='blue', label='Class 1')
    plt.scatter(X[y == 2, 0], X[y == 2, 1], c='purple', label='Class 2')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()
```

```python
def generate_dataset(separability, cluster_sizes):
    np.random.seed(42)

    # Generate Class 0
    class_0 = np.random.randn(cluster_sizes[0], 2) + np.array([0, 0])

    # Generate Class 1 with separability
    class_1 = np.random.randn(cluster_sizes[1], 2) + np.array([separability, separability])

    # Combine data
    X = np.vstack((class_0, class_1))
    y = np.array([0] * cluster_sizes[0] + [1] * cluster_sizes[1])

    return X, y
```

```python
separability_values = [1.0, 2.0, 3.0]
cluster_sizes_values = [(50, 50,50), (100, 50,100), (150, 50,50)]

results = []

for separability in separability_values:
    for cluster_sizes in cluster_sizes_values:
        # Generate dataset
        X, y = generate_dataset(separability, cluster_sizes)
        title = f"Separability={separability}, Cluster Sizes={cluster_sizes}"
        visualize_dataset(X, y, title)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        # Train classifier
        clf = RandomForestClassifier(random_state=42)
        clf.fit(X_train, y_train)

        # Evaluate classifier
        y_pred = clf.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)

        results.append({
            "Separability": separability,
            "Cluster Sizes": cluster_sizes,
            "Accuracy": accuracy
        })

        print(f"Separability: {separability}, Cluster Sizes: {cluster_sizes}")
        print(f"Accuracy: {accuracy:.2f}")
        print(classification_report(y_test, y_pred))

# Display results as a DataFrame
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Sort by Accuracy to find the best configuration
best_results = results_df.sort_values(by="Accuracy", ascending=False)
print("Top Accuracy Results:")
print(best_results.head())
```

```python
import seaborn as sns

# Convert Cluster Sizes to string for categorical plotting
results_df["Cluster Sizes"] = results_df["Cluster Sizes"].astype(str)

# Visualize Accuracy
plt.figure(figsize=(10, 6))
sns.barplot(
    x="Cluster Sizes", 
    y="Accuracy", 
    hue="Separability", 
    data=results_df
)
plt.title("Accuracy Across Cluster Sizes and Separability")
plt.xlabel("Cluster Sizes (e.g., Class 0, Class 1, Class 2, Class 3)")
plt.ylabel("Accuracy")
plt.legend(title="Separability")
plt.xticks(rotation=45)
plt.show()
```
